```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
preprocessor_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
```

```python
bert_preprocess_model = hub.KerasLayer(preprocessor_url)
```

```python
text_test = ["Very nice movie", "That was a shitty film. Waste of time."]
text_preprocessed = bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed["input_mask"]
# it has CLS Very nice movie and SEP. thats why [1, 1, 1, 1, 1] 3 word is 5
```

```python
text_preprocessed["input_word_ids"]
```

```python
bert_model = hub.KerasLayer(encoder_url)
bert_result = bert_model(text_preprocessed)
bert_result.keys()
```
