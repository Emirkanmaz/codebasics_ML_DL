```python
with open('takipci.txt', 'r', encoding="utf-8") as f:
    veriler = f.readlines()
    
takipci = []
for i in range(len(veriler)):
    if "'in profil resmi\n" in veriler[i]:
        takipci.append(veriler[i+1].replace("\n",""))
len(takipci), takipci
```

```python
import pandas as pd
df = pd.DataFrame(takipci)
df.to_excel("takipci_yeni.xlsx", index=False)
df1 = pd.read_excel("takipci_eski.xlsx")
df2 = pd.read_excel("takipci_yeni.xlsx")
pd.concat([df1, df2]).drop_duplicates(keep=False)
```
