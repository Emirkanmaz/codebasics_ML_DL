```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target
```

```python
df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
df
```

```python
df["target"] = digits.target
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(["target"], axis=1),df.target, test_size=0.3 \
                                                   ,random_state=0)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=6)
```

```python
knn_model.fit(X_train, y_train)
knn_model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, knn_model.predict(X_test))
cm
```

```python
import seaborn as sns
import numpy as np

sns.heatmap(cm/np.sum(cm), annot=True, fmt='.2%', cmap='Blues')
```

```python
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import classification_report
```

```python
cf = classification_report(y_test, knn_model.predict(X_test))
print(cf)
```

```python
labels = np.arange(10)
classification_report(y_test, knn_model.predict(X_test), target_names=list("0123456789"), labels=labels, output_dict=True)
```

```python
sns.heatmap(pd.DataFrame(cf).iloc[:-1, :].T, annot=True)
```
