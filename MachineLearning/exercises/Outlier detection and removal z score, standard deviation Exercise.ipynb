You are given bhp.csv which contains property prices in the city of banglore, India. You need to examine price_per_sqft column and do following,

(1) Remove outliers using percentile technique first. Use [0.001, 0.999] for lower and upper bound percentiles

(2) After removing outliers in step 1, you get a new dataframe.

(3) On step(2) dataframe, use 4 standard deviation to remove outliers

(4) Plot histogram for new dataframe that is generated after step (3). Also plot bell curve on same histogram

(5) On step(2) dataframe, use zscore of 4 to remove outliers. This is quite similar to step (3) and you will get exact same result

```python
import pandas as pd
```

```python
df = pd.read_csv("bhp.csv")
df
```

```python
df.price_per_sqft.describe()
```

```python
min_treshold, max_treshold = df["price_per_sqft"].quantile([0.001,0.999])
min_treshold, max_treshold
```

```python
percentile_mask = (df.price_per_sqft>max_treshold) | (df.price_per_sqft<min_treshold)
df[percentile_mask]
```

```python
df2 = df[~percentile_mask]
df2
```

```python
upper_limit = df2.price_per_sqft.mean() + 4 * df2.price_per_sqft.std()
lower_limit = df2.price_per_sqft.mean() - 4 * df2.price_per_sqft.std()
upper_limit, lower_limit
```

```python
std_mask = (df2.price_per_sqft>upper_limit) | (df2.price_per_sqft<lower_limit)
df2[std_mask]
```

```python
df3 = df2[~std_mask]
df3
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8)
plt.show()
```

```python
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8)
plt.yscale('log')
plt.show()
```

```python
df3.price_per_sqft.max(),df3.price_per_sqft.min()
```

```python
from scipy.stats import norm
import numpy as np
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8, density=True)

rng = np.arange(-df3.price_per_sqft.max(), df3.price_per_sqft.max(),0.1)
plt.plot(rng, norm.pdf(rng,df3.price_per_sqft.mean(),df3.price_per_sqft.std()))
plt.show()
```

```python
df2["Z-Score_ppsqft"] = (df2.price_per_sqft - df2.price_per_sqft.mean())/ df2.price_per_sqft.std()
df2
```

```python
zscore_mask = (df2["Z-Score_ppsqft"] > 4) | (df2["Z-Score_ppsqft"] < -4)
df2[zscore_mask]
```

```python
df4 = df2[~zscore_mask]
df4
```

```python
df3==df4
```

```python
df3.equals(df4.drop("Z-Score_ppsqft", axis=1))
```
