```python
df = pd.read_csv("../Bengaluru_House_Data/Bengaluru_House_Data.csv")
df
```

```python
df = df.drop(columns=["area_type", "availability", "society", "balcony"])
df
```

```python
def convert_sqft_to_num(x):
    tokens = x.split("-")
    if len(tokens) == 2:
        return np.mean([float(tokens[0]), float(tokens[1])])
    try:
        return float(x)
    except:
        pass    
```

```python
df["total_sqft"] = df["total_sqft"].apply(convert_sqft_to_num)
df
```

```python
df.dropna(subset=['size'], inplace=True)
```

```python
df["bhk"] = df["size"].apply(lambda x: x.split(" ")[0])
```

```python
df
```

```python
df["price_per_sqft"] = df["price"]*100_000/df["total_sqft"]
df
```
