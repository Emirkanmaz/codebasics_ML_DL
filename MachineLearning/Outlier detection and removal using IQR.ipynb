```python
IQR = df.height.quantile(.75) - df.height.quantile(.25)
lower_limit = df.height.quantile(.25) - IQR 
upper_limit = df.height.quantile(.75) + IQR 
IQR, upper_limit, lower_limit
```

```python
IQR_mask = (df.height>upper_limit) | (df.height<lower_limit)
df[IQR_mask]
```
