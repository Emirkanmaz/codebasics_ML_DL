```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import svm
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
df["flower"] = iris.target
df["flower"]
```

```python
df["flower"] = df.flower.apply(lambda x: iris.target_names[x])
df
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(svm.SVC(gamma="auto"), {
    "C": [1,10,20],
    "kernel": ["rbf", "linear"]
}, cv=5, return_train_score=False)
clf.fit(iris.data, iris.target)
```

```python
dir(clf)
```

```python
clf.cv_results_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rs = RandomizedSearchCV(svm.SVC(gamma="auto"), {
    "C": [1,10,20],
    "kernel": ["rbf", "linear", "sigmoid", "poly"]
}, cv=5, return_train_score=False, n_iter=5)

rs.fit(iris.data, iris.target)
pd.DataFrame(rs.cv_results_)[["param_C", "param_kernel", "mean_test_score"]]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    "svm":{
        "model":svm.SVC(gamma="auto"),
        "params":{
            "C": [1,10,20],
            "kernel": ["rbf", "linear", "sigmoid", "poly"]
        } 
    },
    "random_forest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10,25,50,100]
        }
    },
    "logistic_regression":{
        "model": LogisticRegression(),
        "params":{
            "solver":['lbfgs','liblinear'],
            "C":[1,5,10]
        }
    }
}
```

```python
scores = []

for model_name, model_pointer in model_params.items():
    clf = GridSearchCV(model_pointer["model"], model_pointer["params"], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append({
        "model": model_name,
        "best_score":clf.best_score_,
        "best_params":clf.best_params_
    })
```

```python
df = pd.DataFrame(scores)
df
```
