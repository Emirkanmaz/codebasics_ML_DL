# Titanic Naive Bayes

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv", index_col="PassengerId", usecols=["PassengerId", "Sex", "Age", "Pclass", "Fare", "Survived"])
df
```

```python
y=df.Survived
x=df.drop("Survived", axis=1)
x
```

```python
dummies = pd.get_dummies(x["Sex"])
dummies
```

```python
x = pd.concat([x, dummies], axis=1)
x
```

```python
x.drop(["Sex", "male"], axis=1, inplace=True)
x
```

```python
x[x.Age.isna()]
```

```python
x["Age"].fillna(x["Age"].mean(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
NB_model = GaussianNB()
NB_model.fit(X_train, y_train)
```

```python
NB_model.score(X_test, y_test)
```

```python
NB_model.predict(X_test[:10])
```

```python
y_test[:10]
```

```python
NB_model.predict_proba(X_test[:10])
```

# Spam Naive Bayes 

```python
import pandas as pd
```

```python
df = pd.read_csv("spam.csv")
df
```

```python
df["Spam"] = pd.get_dummies(df.Category).spam
df
```

```python
df.drop(["Category"], axis=1, inplace=True)
df
```

```python
df.isna().any()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.Spam, test_size=0.2, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
Bernoulli_model = BernoulliNB()
Multinomial_model = MultinomialNB()
```

```python
Bernoulli_model.fit(X_train_count, y_train)
Multinomial_model.fit(X_train_count,y_train)
```

```python
X_test_count = v.transform(X_test)
```

```python
Bernoulli_model.score(X_test_count, y_test)
```

```python
Multinomial_model.score(X_test_count, y_test)
```

# Pipeline

```python
from sklearn.pipeline import Pipeline
Bernoulli_clf = Pipeline([
    ("Vectorizer", CountVectorizer()),
    ("BernoulliNB", BernoulliNB())
])
```

```python
Multinomial_clf = Pipeline([
    ("Vectorizer", CountVectorizer()),
    ("MultinomialNB", MultinomialNB())
])
```

```python
Bernoulli_clf.fit(X_train,y_train)
Bernoulli_clf.score(X_test,y_test)
```

```python
Multinomial_clf.fit(X_train,y_train)
Multinomial_clf.score(X_test,y_test)
```
